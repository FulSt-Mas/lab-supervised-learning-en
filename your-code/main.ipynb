# Challenge 3 - Handle Missing Values

The next step would be handling missing values. **We start by examining the number of missing values in each column, which you will do in the next cell.**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
websites = pd.read_csv('../website.csv')

# Initial numeric column selection
numeric_cols = websites.select_dtypes(include=['float64', 'int64'])

# Function to compute and plot correlation matrix
def plot_correlation_matrix(df):
    corr_matrix = df.corr()
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Correlation Matrix of Numeric Features')
    plt.show()
    return corr_matrix

# Identify and remove columns with high collinearity
def remove_highly_correlated(df, threshold=0.9):
    corr_matrix = df.corr().abs()
    upper_triangle = corr_matrix.where(~np.tril(np.ones(corr_matrix.shape)).astype(bool))

    # Find column pairs with a correlation coefficient above the threshold
    to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)]

    if to_drop:
        print(f"\nColumns to drop (correlation above {threshold*100}%): {to_drop}")
        df.drop(columns=to_drop, inplace=True)
    else:
        print("No columns with high collinearity remaining.")

    return df

# Iterate until no highly correlated columns remain
while True:
    plot_correlation_matrix(numeric_cols)
    new_numeric_cols = remove_highly_correlated(numeric_cols)

    # Break if no columns were dropped
    if numeric_cols.equals(new_numeric_cols):
        break

    numeric_cols = new_numeric_cols

# Final heatmap
print("\nFinal heatmap:")
plot_correlation_matrix(numeric_cols)
```

There are several categorical columns we need to handle. These columns are:

How to handle string columns is always case by case. Let's start by working on `WHOIS_COUNTRY`. Your steps are:

1. List out the unique values of `WHOIS_COUNTRY`.
1. Consolidate the country values with consistent country codes. For example, the following values refer to the same country and should use consistent country code:
    * `CY` and `Cyprus`
    * `US` and `us`
    * `SE` and `se`
    * `GB`, `United Kingdom`, and `[u'GB'; u'UK']`

#### In the cells below, fix the country values as intructed above.

```python
import pandas as pd

# Load the dataset
websites = pd.read_csv('../website.csv')

# List unique values in `WHOIS_COUNTRY`
unique_countries = websites['WHOIS_COUNTRY'].unique()
print("Unique values in `WHOIS_COUNTRY`:\n", unique_countries)

# Define a mapping from inconsistent representations to standardized country codes
country_mapping = {
    'CY': 'CY', 'Cyprus': 'CY',
    'US': 'US', 'us': 'US',
    'SE': 'SE', 'se': 'SE',
    'GB': 'GB', 'United Kingdom': 'GB', "[u'GB'; u'UK']": 'GB'
}

# Apply the mapping to standardize the `WHOIS_COUNTRY` values
websites['WHOIS_COUNTRY'] = websites['WHOIS_COUNTRY'].replace(country_mapping)

# Verify the consolidation
updated_unique_countries = websites['WHOIS_COUNTRY'].unique()
print("\nUpdated unique values in `WHOIS_COUNTRY`:\n", updated_unique_countries)
```

Since we have fixed the country values, can we convert this column to ordinal now?

Not yet. If you reflect on the previous labs how we handle categorical columns, you probably remember we ended up dropping a lot of those columns because there are too many unique values. Too many unique values in a column is not desirable in machine learning because it makes prediction inaccurate. But there are workarounds under certain conditions. One of the fixable conditions is:

#### If a limited number of values account for the majority of data, we can retain these top values and re-label all other rare values.

The `WHOIS_COUNTRY` column happens to be this case. You can verify it by print a bar chart of the `value_counts` in the next cell to verify:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
websites = pd.read_csv('../website.csv')

# Generate value counts for `WHOIS_COUNTRY`
country_counts = websites['WHOIS_COUNTRY'].value_counts()

# Plot a bar chart of the value counts to visualize
plt.figure(figsize=(12, 6))
country_counts.plot(kind='bar')
plt.title('Value Counts of WHOIS_COUNTRY')
plt.xlabel('Country Code')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Define the threshold for keeping a country
threshold_percentage = 0.90

# Calculate cumulative distribution percentage
cumulative_distribution = country_counts.cumsum() / country_counts.sum()

# Identify countries to retain
countries_to_keep = cumulative_distribution[cumulative_distribution <= threshold_percentage].index

# Transform the column: keep frequent, others become 'Other'
websites['WHOIS_COUNTRY'] = websites['WHOIS_COUNTRY'].apply(lambda x: x if x in countries_to_keep else 'Other')

# Check the updated value counts
print(websites['WHOIS_COUNTRY'].value_counts())
```

Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`. Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
import pandas as pd

# Load your cleaned dataset
websites_cleaned = pd.read_csv('../website.csv')

# Drop previous columns for a cleaner dataset
columns_to_drop = ['URL', 'WHOIS_STATEPRO', 'WHOIS_REGDATE', 'WHOIS_UPDATED_DATE']
websites_cleaned = websites_cleaned.drop(columns=columns_to_drop)

# Define a function to classify server types
def classify_server(server_name):
    if isinstance(server_name, str):  # Check if the value is a string
        server_name_lower = server_name.lower()
        if 'microsoft' in server_name_lower:
            return 'Microsoft'
        elif 'apache' in server_name_lower:
            return 'Apache'
        elif 'nginx' in server_name_lower:
            return 'nginx'
    return 'Other'  # Default to 'Other' if not a string or doesn't match

# Apply the function to the SERVER column to classify servers
websites_cleaned['SERVER'] = websites_cleaned['SERVER'].apply(classify_server)

# Verify the transformation by printing the unique values
print("Transformed unique values in SERVER column:\n", websites_cleaned['SERVER'].unique())
```

This step needs to happen during Challenge 6 after you split the training and test data because you don't want to split the data again which makes it impossible to compare your results with and without feature scaling. For general concepts about feature scaling, click [here](https://en.wikipedia.org/wiki/Feature_scaling). To read deeper, click [here](https://medium.com/greyatom/why-how-and-when-to-scale-your-features-4b30ab09db5e).

In the next cell, attempt to improve your model prediction accuracy by means of feature scaling. A library you can utilize is `sklearn.preprocessing.RobustScaler` ([documentation](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html)). You'll use the `RobustScaler` to fit and transform your `X_train`, then transform `X_test`. You will use logistic regression to fit and predict your transformed data and obtain the accuracy score in the same way. Compare the accuracy score with your normalized data with the previous accuracy data. Is there an improvement?

```python
# Your code here

from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Initialize the RobustScaler
scaler = RobustScaler()

# Fit the scaler on the training data and transform both train and test sets
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# Initialize the logistic regression model
logistic_model_scaled = LogisticRegression(max_iter=1000, random_state=42)

# Fit the logistic regression model to the scaled training data
logistic_model_scaled.fit(X_train_scaled, y_train)

# Predict the test set results
y_pred_scaled = logistic_model_scaled.predict(X_test_scaled)

# Compute and print the confusion matrix of the scaled model
conf_matrix_scaled = confusion_matrix(y_test, y_pred_scaled)
print("Confusion Matrix with Scaled Features:\n", conf_matrix_scaled)

# Compute and print the accuracy score of the scaled model
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print("Accuracy Score with Scaled Features:", accuracy_scaled)
```
